```python
from google_play_scraper import reviews, Sort
import pandas as pd
import time
import os
```

```python
# App configuration
APP_ID = 'com.olaelectric.companion'
COUNTRY = 'in'        # India
LANGUAGE = 'en'       # English reviews only for now

# Scraping configuration
TOTAL_REVIEWS = 5000
BATCH_SIZE = 200      # Pull 200 at a time — avoids rate limiting

# Output configuration
OUTPUT_PATH = r'C:\Users\Naren karthikeyan\OLA Analysis\data\raw\ola_reviews_raw.csv'
```

```python
def scrape_ola_reviews(app_id, total, batch_size, country, language):
    """
    Scrapes reviews from Google Play Store in batches.
    Returns a clean dataframe.
    """
    all_reviews = []
    continuation_token = None
    batch_number = 1

    print(f"Starting scrape for: {app_id}")
    print(f"Target: {total} reviews | Batch size: {batch_size}")
    print("-" * 50)

    while len(all_reviews) < total:
        try:
            # Calculate how many to fetch this batch
            remaining = total - len(all_reviews)
            fetch_count = min(batch_size, remaining)

            # Fetch the batch
            batch, continuation_token = reviews(
                app_id,
                lang=language,
                country=country,
                sort=Sort.NEWEST,        # Most recent first
                count=fetch_count,
                continuation_token=continuation_token
            )

            # If no more reviews available, stop
            if not batch:
                print(f"No more reviews available. Stopping at {len(all_reviews)} reviews.")
                break

            all_reviews.extend(batch)

            print(f"Batch {batch_number} complete — {len(all_reviews)} reviews collected so far")

            batch_number += 1

            # Polite pause between batches — avoids getting blocked
            time.sleep(2)

            # If no continuation token, we've hit the end
            if not continuation_token:
                print("Reached end of available reviews.")
                break

        except Exception as e:
            print(f"Error on batch {batch_number}: {e}")
            print("Waiting 10 seconds before retrying...")
            time.sleep(10)
            continue

    print("-" * 50)
    print(f"Scraping complete. Total reviews collected: {len(all_reviews)}")
    return all_reviews
```

```python
raw_reviews = scrape_ola_reviews(
    app_id=APP_ID,
    total=TOTAL_REVIEWS,
    batch_size=BATCH_SIZE,
    country=COUNTRY,
    language=LANGUAGE
)
```

```python
df_raw = pd.DataFrame(raw_reviews)

# Preview what columns we got
print("Columns available:")
print(df_raw.columns.tolist())
print(f"\nShape: {df_raw.shape}")
print("\nFirst review:")
df_raw.head(1)
```

```python
print("-" * 50)
print("SANITY CHECK REPORT")
print("-" * 50)

# 1. Row count
print(f"\nTotal reviews: {len(df_raw)}")

# 2. Date range
print(f"Date range: {df_raw['at'].min()} → {df_raw['at'].max()}")

# 3. Rating distribution
print(f"\nRating distribution:")
print(df_raw['score'].value_counts().sort_index())

# 4. Null check
print(f"\nNull values per column:")
print(df_raw.isnull().sum())

# 5. Duplicate check
print(f"\nDuplicate reviews: {df_raw.duplicated(subset='reviewId').sum()}")

# 6. OLA reply rate
reply_rate = df_raw['replyContent'].notna().mean() * 100
print(f"\nOLA reply rate: {reply_rate:.1f}%")
```

```python
# Save
df_raw.to_csv(OUTPUT_PATH, index=False)

print(f"Raw data saved to: {OUTPUT_PATH}")
print(f"File size: {os.path.getsize(OUTPUT_PATH) / 1024:.1f} KB")
```

```python
from google_play_scraper import Sort
import time

def scrape_historical_reviews(app_id, total, batch_size, country, language):
    """
    Pulls historically significant reviews using MOST_RELEVANT sort.
    Captures older high-engagement reviews missed by NEWEST sort.
    """
    all_reviews = []
    continuation_token = None
    batch_number = 1

    print(f"Starting HISTORICAL scrape for: {app_id}")
    print(f"Target: {total} reviews | Sort: MOST_RELEVANT")
    print("-" * 50)

    while len(all_reviews) < total:
        try:
            remaining = total - len(all_reviews)
            fetch_count = min(batch_size, remaining)

            batch, continuation_token = reviews(
                app_id,
                lang=language,
                country=country,
                sort=Sort.MOST_RELEVANT,   # Key difference
                count=fetch_count,
                continuation_token=continuation_token
            )

            if not batch:
                print(f"No more reviews available. Stopping at {len(all_reviews)}.")
                break

            all_reviews.extend(batch)
            print(f"Batch {batch_number} complete — {len(all_reviews)} reviews so far")

            batch_number += 1
            time.sleep(2)

            if not continuation_token:
                print("Reached end of available reviews.")
                break

        except Exception as e:
            print(f"Error on batch {batch_number}: {e}")
            time.sleep(10)
            continue

    print("-" * 50)
    print(f"Historical scrape complete. Total: {len(all_reviews)}")
    return all_reviews

# Run it
raw_reviews_historical = scrape_historical_reviews(
    app_id=APP_ID,
    total=3000,          # 3000 additional reviews
    batch_size=200,
    country=COUNTRY,
    language=LANGUAGE
)
```

```python
df_historical = pd.DataFrame(raw_reviews_historical)

print(f"Historical batch shape: {df_historical.shape}")
print(f"Date range: {df_historical['at'].min()} → {df_historical['at'].max()}")
print(f"\nRating distribution:")
print(df_historical['score'].value_counts().sort_index())
```

```python
# Load your original saved dataset
df_newest = pd.read_csv(r'C:\Users\Naren karthikeyan\OLA Analysis\data\raw\ola_reviews_raw.csv')

print(f"Newest batch: {len(df_newest)} reviews")
print(f"Historical batch: {len(df_historical)} reviews")

# Combine
df_combined = pd.concat([df_newest, df_historical], ignore_index=True)

print(f"\nCombined total before dedup: {len(df_combined)}")
```

```python
# Remove duplicates based on unique review ID
df_combined = df_combined.drop_duplicates(subset='reviewId', keep='first')

print(f"Combined total after dedup: {len(df_combined)}")
print(f"Duplicates removed: {(len(df_newest) + len(df_historical)) - len(df_combined)}")
```
